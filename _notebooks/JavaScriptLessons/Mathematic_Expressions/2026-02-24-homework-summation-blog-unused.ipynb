```python
%%js

// CODE_RUNNER: Exercise 2: Modify Arrays

// Start with the shopping list
let shoppingList = ["milk", "eggs", "bread", "cheese"];

// Print the original array
console.log("Original array:", shoppingList);

// Change the second item (index 1) to "butter"
shoppingList[1] = "butter";

// Add "yogurt" to the end using push()
shoppingList.push("yogurt");

// Remove "bread" from the array
// Option 1: Using splice()
const breadIndex = shoppingList.indexOf("bread");
if (breadIndex !== -1) {
    shoppingList.splice(breadIndex, 1);
}

// Print the final array
console.log("Final array:", shoppingList);
```

```python
%%js

// CODE_RUNNER: Exercise 3: Loop Through and Array

// Create the array
const numbers = [10, 25, 30, 15, 20];

// Variable to keep track of the sum
let sum = 0;

// Loop through the array
for (let i = 0; i < numbers.length; i++) {
    const num = numbers[i];

    // Print each number with a message
    console.log("Number:", num);

    // Print each number multiplied by 2
    console.log("Number x 2:", num * 2);

    // Add to the sum
    sum += num;
}

// Print the total sum
console.log("Total sum:", sum);
```

```python
%%js

// CODE_RUNNER: Problem 1

function calculator(num1, num2, operator) {
    let result;

    if (operator === "+") {
        result = num1 + num2;
    } else if (operator === "-") {
        result = num1 - num2;
    } else if (operator === "*") {
        result = num1 * num2;
    } else if (operator === "/") {
        result = num1 / num2;
    } else {
        result = "Invalid operator";
    }

    return result;
}

console.log(calculator(10, 5, "+"));
console.log(calculator(10, 5, "-"));
console.log(calculator(10, 5, "*"));
console.log(calculator(10, 5, "/"));
```

```python
%%js

// CODE_RUNNER: Problem 1

// Factors of 50
const factors = [1, 2, 5, 10, 25, 50];

for (let num = 1; num <= 50; num++) {

    // First nested conditional: check divisibility by 1
    if (num % factors[0] === 0) {

        // Second nested conditional: check divisibility by 2
        if (num % factors[1] === 0) {

            // Third nested conditional: check divisibility by 5
            if (num % factors[2] === 0) {

                // Fourth nested conditional: check divisibility by 10
                if (num % factors[3] === 0) {

                    // Fifth nested conditional: check divisibility by 25
                    if (num % factors[4] === 0) {

                        // Sixth nested conditional: check divisibility by 50
                        if (num % factors[5] === 0) {
                            console.log(num + " is divisible by ALL factors of 50");
                        } else {
                            console.log(num + " is divisible by 1,2,5,10,25 but NOT 50");
                        }

                    } else {
                        console.log(num + " is divisible by 1,2,5,10 but NOT 25");
                    }

                } else {
                    console.log(num + " is divisible by 1,2,5 but NOT 10");
                }

            } else {
                console.log(num + " is divisible by 1,2 but NOT 5");
            }

        } else {
            console.log(num + " is divisible by 1 but NOT 2");
        }

    } else {
        console.log(num + " is NOT divisible by 1 (this never happens)");
    }
}
```
